```python
from faker import Faker
from pathlib import Path
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
hdfc_rows = []

fake = Faker()
today = datetime.now()+ timedelta(days=1)
vendor = "HDFC"

# Create output folder if it doesn't exist
output_folder = Path(
    f"../../data/raw/{vendor}/year={today.year}/month={today.month:02d}/day={today.day:02d}"
)

output_folder.mkdir(parents=True, exist_ok=True)


funds = ["H101", "H102", "H103", "H104"]


for i in range(100):
    hdfc_rows.append({
        "TradeDate": today.date(),
        "InvestorCode": f"C{1000+i}",
        "FundCode": random.choice(funds),
        "TxnType": random.choice(["PURCHASE", "REDEMPTION"]),
        "Amount": random.randint(1000,50000),
        "Units": random.randint(10,500)
    })
# Build full file path

df = pd.DataFrame(hdfc_rows)

file_name = (
    f"HCFC_transactions_"
    f"{today.strftime('%Y%m%d_%H%M%S')}.csv"
)

# Build full file path
output_file = output_folder / file_name
df.to_csv(output_file, index=False)

print(f"File saved to: {output_file}")
```
