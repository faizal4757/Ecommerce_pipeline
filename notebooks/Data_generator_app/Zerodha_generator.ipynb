```python
from faker import Faker
from pathlib import Path
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
zerodha_rows = []
fake = Faker()
today = datetime.now() + timedelta(days=1)
vendor = "ZERODHA"

# Create output folder if it doesn't exist
output_folder = Path(
    f"../../data/raw/{vendor}/year={today.year}/month={today.month:02d}/day={today.day:02d}"
)

output_folder.mkdir(parents=True, exist_ok=True)

funds = ["F101", "F102", "F103", "F104"]

for i in range(100):
    zerodha_rows.append({
        "transaction_date": today.date(),
        "user_id": f"Z{7000+i}",
        "scheme_code": random.choice(funds),
        "txn_flag": random.choice(["B","S"]),
        "investment_amount": random.randint(1000,50000),
        "unit_count": random.randint(10,500)
    })

df = pd.DataFrame(zerodha_rows)

file_name = (
    f"zerodha_transactions_"
    f"{today.strftime('%Y%m%d_%H%M%S')}.csv"
)

# Build full file path
output_file = output_folder / file_name
df.to_csv(output_file, index=False)

print(f"File saved to: {output_file}")
```
